# The Law of Mass Action

The first topic we cover is the *law of mass action*, which models the rate of chemical reactions and is one of the most fundamental concepts in chemistry. As an example we will look at the binding reaction between oxygen and *myoglobin*, an oxygen-binding globular protein found in muscle cells. If an environment contains both free oxygen molecules and myoglobin, the two can combine to form a *oxymyoglobin* complex. We can write this reaction as the following chemical equation 
\begin{equation}
\label{eq:myoglobin_one_way}
\mathrm{Mb} + \mathrm{O_2} \overset{k}{\longrightarrow} \mathrm{MbO_2}.
\end{equation}

This equation describes the binding reaction where a free myoglobin (Mb) and a oxygen molecule (O$_2$) combine to form a oxymyoglobin (MbO$_2$). This reaction will occur spontaneously, but not instantaneously. Instead it will progress at a certain *rate*, and this is where the law of mass action comes in.

The main idea behind the law of mass action is that for a binding reaction to occur at all, the two reacting molecules need to be close enough to each other to interact. At the molecular level, the reactants are not stationary, but moving around at high speeds in a chaotic and random manner called thermal motion.

Due to this motion, many meetings between the reactants will occur. Not every such meeting will results in a binding reaction, but the rate of the reaction at the macroscopic level will therefore be proportional to the number of meetings. But how often do such meetings occur? Because thermal motion is inherently random, it is very reasonable to assume that the number of meetings will be proportional to the concentrations of the reactants. The more molecules available, the more meetings will occur. If we use the square brackets to denote the concentrations, we can write this as

\begin{equation}
\text{rate} \propto [\mathrm{Mb}]\times[\mathrm{O_2}].
\end{equation}

Thus the *rate* of the binding reaction is proportional to the concentration of the reactants.

The proportionality constant for the reaction is typically called the *rate constant* and denoted $k$, which is why we have written this letter the arrow in the chemical equation. The rate constant is related to the probability of the reaction occurring when two such molecules meet, known as the *reaction cross-section*. The higher the chance the molecules have of reacting, the higher the rate constant. The rate constant will therefore be different for different reactions. However, whatever the rate constant, the rate of the reaction will be proportional to the concentrations of the reactants.

With the rate of the reaction derived, we can set up a system of ordinary differential equations (ODEs) describing the time evolution of the different concentrations. As we have three molecular species, the two reactants and the one product, we get three ODEs. These become

\begin{equation}
\frac{\mathrm{d[MbO_2]}}{\mathrm{d}t} = k\mathrm{[Mb][O_2]}, \qquad \frac{\mathrm{d [Mb]}}{\mathrm{d}t} = -k\mathrm{[Mb]}\mathrm{[O_2]}, \qquad \frac{\mathrm{d [O_2]}}{\mathrm{d}t} = -k\mathrm{[Mb]}\mathrm{[O_2]}.
\end{equation}

This system of ODEs is a mathematical model of the binding reaction and we can solve these to find how the concentrations changes over time. To do so we would need a specific value for the rate constant $k$, which we would measure in an empirical experiment. We would also need initial conditions for the three reactions to form a complete initial value problem we can solve.

An elementary reaction is any reaction where the reactants combine and form a product in a single step. Not all reactions are elementary, but more complex reactions can be described as a series of elementary reactions. Modeling reactions in this manner using the law of mass action is therefore a general and powerful technique. 

Note also that while we state that the rate is proportional to the concentrations of the reactants, it would be more accurate to say it is proportional to their *activities*. The activity of a given molecular species can be described as the *effective concentration* of a species and it is generally lower than the actual concentration in a more complex solution that do not behave like *ideal solutions* were the reactants are free to move around independently of each other. The details are not that important for our current discussion.

#### Determining and scaling rate constants

To model chemical reactions with the law of mass action we need to know the rate constants and these must be found empirically. Even though myoglobin is a large and complex macromolecule, its functional behavior however is quite simple to model with the law of mass action as long as we measure its rate constant in an experiment. 

We call them rate *constants* because they do not change with the concentrations of the reactants, note however that they can change for other reasons. If the temperature is increased for example the rate of a reaction typically increases as well, and this is captured in the model by increasing the rate constant. The rate constant also depends on other environmental factors such as pressure and pH.

In this course we are mostly looking at closed biological systems where these factors can be considered constant. Therefore, we generally do not need to incorporate temperature or pressure driven shifts in rates in our models. However, it is important to be aware that many experiments carried out in a lab are not carried out as the same temperature as those seen *in situ*. An experiment might be carried out in room temperature (~20$^\circ$C), but then you want to use it for a model in physiological conditions (~37$^\circ$C). In this case you would need to scale the rate constant to account for the temperature shift, this is typically done with a temperature coefficient, denoted $Q_{\rm 10}$, which tells you how much a given model parameter needs to be scaled.

#### Activation energy and the Arrhenius equation

To understand why the reaction rate increases with temperature, we can turn to the *Arrhenius equation*, which states that a general rate constant can be expressed as

\begin{equation}
k = Ae^{-E_{\rm a}/RT}.
\end{equation}

Here $k$ is a rate constant, $A$ is called the pre-exponential factor which is species-dependent but temperature independent, we then have an exponential term where $E_{\rm a}$ represents the *activation energy* of the reaction, $R$ the ideal gas constant and $T$ the temperature (in absolute units).

Arrhenius' equation tells us that the rate constant has an exponential dependence on temperature, and as the temperature is lowered, the rate constant is dampened. The activation energy $E_{\rm a}$ represents the energy the reactants need to overcome to actually react, and it is related to the reaction cross-section described earlier. When the temperature increases, the thermal motion of the reactants is energized and so it will be easier to overcome the activation barrier and actually react.

## Reversible reactions

When formulating our model for the binding reaction, we made an implicit assumption that the *reverse* reaction doesn't occur, i.e., an oxymyoglobin splitting into an oxygen molecule and a myoglobin, a process we call *dissociation*. Many chemical reactions are reversible, and will occur in both directions, especially so for a closed system such as the inside of a cell. The binding and dissociation of oxygen will occur in parallel and we therefore modify our chemical equation to reflect this
\begin{equation}
\mathrm{Mb} + \mathrm{O_2} \underset{k_-}{\overset{k_+}{\rightleftharpoons}} \mathrm{MbO_2}.
\end{equation}
Here we have replaced the unidirectional arrow with a bi-directional pair of arrows representing both the forward and reverse directions of the reaction. Note also that these have been given different rate constants, and we denote these $k_+$ and $k_-$ to tell them apart.

The rate of the forward reaction will be the same as before, and the rate of the reverse reaction can be formulated as

\begin{equation}
\text{reverse rate} = k_-\mathrm{[MbO_2]}
\end{equation}

The rate is again found by assuming the rate to be proportional to the reactants concentrations, and this is true even if the reactant is a single molecule.

In an environment containing both oxygen and myoglobin we will now have a complex dynamic situation where myoglobin and oyxgen are continuously binding together and dissociating. The *net* change in concentrations over time can be found by looking at the difference between the forward and reverse rate. The ODE describing the concentration of oxymyoglobin will for example be given  by

If we now want to formulate the change in oxymgoblin over time, 

\begin{equation}
\frac{\mathrm{d[MbO}_2]}{\mathrm{d}t} = k_+[\mathrm{Mb}][\mathrm{O_2}] - k_-[\mathrm{MbO_2}].
\end{equation}

The first term corresponds to the forward rate and is positive, because oxymyoglobin is formed when a binding reaction occurs. The second term is the reverse rate, which we subtract because oxymyoglobin are removed from the system when they dissociate. 

## Chemical Equilibrium

When modeling a unidirectional chemical reaction, the concentration of the reactants can only drop over time while the concentrations of the product can only increase. When we look at a reversible reaction however, the concentrations of the different molecules can both increase and decrease. Looking at ODE we just formulated for example, we see the concentration of oxymyoglobin increases if the forward rate is larger than the reverse, and decreases if the situation is flipped. As the concentration changes over time however, these two rates will also change and approach each other until we end up in a chemical *equilibrium*.

An equilibrium is reached once the forward and reverse rate of the reaction are of equal magnitude and effectively cancel each other out. We can write this as the following equilibrium criteria

\begin{equation}
\frac{\mathrm{d[MbO}_2]}{\mathrm{d} t} = k_+[\mathrm{Mb}][\mathrm{O_2}] - k_-[\mathrm{MbO_2}] = 0 \quad (\text{at equilibrium}).
\end{equation}

This criteria does not say that the reaction has stopped, but rather that the forward and reverse reactions are now occurring at the same rate, meaning there is no *net* change in concentrations. This is referred to as a *dynamic* equilibrium, to emphasize that the reaction is still occurring at the molecular level.

The equilibrium condition states that the two rates are equal

\begin{equation}
    k_+[\mathrm{Mb}][\mathrm{O_2}] = k_-[\mathrm{MbO_2}] \quad (\text{at equilibrium}).
\end{equation}

And from this criteria we can see that the *ratio* of the reactants and the product must be constant at equilibrium

\begin{equation}
\label{eq:equilibrium_equation}
\frac{[\mathrm{Mb}][\mathrm{O_2}]}{[\mathrm{MbO_2}]} = \frac{k_-}{k_+} \equiv K_{\rm d} \quad (\text{at equilibrium}).
\end{equation}
The ratio of the rate constants is typically referred to as the *dissociation constant* of the reaction, and it is denoted $K_{\rm d}$. The higher the dissociation constant, the higher the concentrations of the free molecules will tend to be at equilibrium. Some refer to the dissociation constant simply as the *equilibrium* constant, $K_{\rm eq}$. Note however that the term equilibrium doesn't indicate which direction of the reaction we are referring and so this term could also mean the inverse relation $k_+/k_-$, which would be the *association* constant. There are no strict conventions of which of these are used, and some care should be taken to interpret how the equilibrium constant for a reaction is formulated and used.

This equation is referred to as *the equilibrium equation* and it describes how the ratio of concentrations will settle at equilibrium. It is a direct consequence of the law of mass action. If one is referring to the law of mass action as an equation, it is typically the equilibrium equation that is meant. Note that the equilibrium equation does not hold at all times, but only when a system is at equilibrium. However, any chemical system will spontaneously tend towards this state, and for many reactions this occurs so swiftly that one can consider the system to always be at equilibrium for all practical purposes.

The equilibrium equation is very useful, but it might be a bit challenging to see how this equation is interpreted or used in practice. Let us therefore look at an illustrative example.

#### Fractional binding

Myoglobin is found in muscle cells and in this environment we can think of the total amount of myoglobin as fixed, i.e.,
$$[\mathrm{Mb}]_{\rm tot} = [\mathrm{Mb}] + [\mathrm{MbO_2}] = \mathrm{const.}$$
The total amount of oxygen in the cell however, can change over time. Oxygen is supplied by the circulatory system and is free to diffuse into the cell. When an oxygen binds to myoglobin, more oxygen will spontaneously, and so the concentration of free oxygen $[\mathrm{O}_2]$ can be thought of as constant within the cell, at least over shorter time scales.

We now introduce the *fractional degree of oxygenation*, i.e., the fraction of all myoglobin molecules that will have oxygen bound to them. This corresponds to the following fraction

\begin{equation}
Y = \frac{[\text{bound Mb}]}{[\text{total Mb}]} = \frac{[\mathrm{MbO_2}]}{[\mathrm{Mb}] + [\mathrm{MbO_2}]}.
\end{equation}

We will now use the equilibrium equation to find an expression of $Y$ as a functional of the free oxygen concentration. To simplify this derivation, we are only concerned with the fractional oxygenation at equilibrium. We start by solving the equilibrium equation for $[\mathrm{MbO_2}]$, giving

\begin{equation}
[\mathrm{MbO_2}] = K_{\rm d}^{-1}[\mathrm{Mb}][\mathrm{O_2}].
\end{equation}

We can now insert this into our expression for $Y$, yielding

\begin{equation}
Y = \frac{K_{\rm d}^{-1}[\mathrm{Mb}][\mathrm{O_2}]}{[\mathrm{Mb}] + K_{\rm d}^{-1}[\mathrm{Mb}][\mathrm{O_2}]}.
\end{equation}

Simplifying, we find a much neater expression:

\begin{equation}
\label{eq:myoglobin}
Y = \frac{[\mathrm{O_2}]}{[\mathrm{O_2}] + K_{\rm d}}.
\end{equation}

To summarize, this expression describes what ratio of myoglobin molecules will have oxygen bound to them as a function of the concentration of free oxygen. We graph this expression in the following figure.

<img src="fig/myoglobin.png" width=500>
<center><b>Figure 2:</b> The fractional degree of oxygenation of myoglobin as a function of free oxygen, which we have scaled by the dissociation constant $K_{\rm d}$.</center>

The curve tells us the fraction of myoglobin proteins that have oxygen bound to it at different levels of free oxygen. If there is no oxygen available, no myoglobin can bind oxygen either. As some oxygen is added to the system some myoglobin will start binding it. If the concentration of free oxygen is low, the fractional oxygenation will also be low, regardless of how much oxygen is available *in total*, it is the concentration of oxygen that is important, due to the law of mass action. As the concentration of free oxygen increases, so does the degree of oxygenation. Note especially that when $[{\rm O}_2] = K_{\rm d}$, we get $Y=0.5$. In this scenario, the dissociation constant is equal to the point of half-saturation of the system. As the oxygen levels rise higher than this the oxygenation rises slower and slower, and you can see the oxygen concentrations need to be very high for the oxygenation of myoglobin to approach 100\%.

In this example we have derived the fractional oxygenation of myoglobin using the law of mass action and the equilibrium equation. This might seem like a very specific example, especially due to our assumption of a constant level of free oxygen. However, all derivations we have done come directly from the law of mass action and are therefore very general. The assumption of a constant amount of free oxygen can also be a good approximation of a system where there is plentiful of one reactant, and comparatively less of the other. In the live session we will look closer at enzyme kinematics, which we will model as such a system.

## Modeling Cooperativity and The Hill Equation

We now turn to an important phenomenon in biophysical systems known as *cooperativity*, which in turn will lead us to the *Hill equation*. The example we use to derive the Hill equation will mirror the derivation we just carried out, and will again rely directly on the law of mass action.

We will now introduce a different oxygen-binding protein called *hemoglobin*. When a myoglobin protein can bind a single oxygen molecule, a hemoglobin can bind up to four oxygen molecules at once. Put simply, a hemoglobin molecule is a lot like four individual myoglobin proteins stapled together.

To model how hemoglobin binds oxygen we can envision each hemoglobin to be in one of five states, ranging from having zero oxygen molecules bound up to having all four bonding sites occupied. As illustrated in the following figure

<img src="fig/hemoglobin_reaction.png" width=600>

<center><b>Figure 3:</b> A single hemoglobin molecule has four binding sites for oxygen.</center>

To properly model this system we would need to set up six ODEs, one for each species in the system. However, we can simplify the system considerably. If it were the case that each of these binding and dissociation reactions had the same rate constants, we could simplify this system considerably and say that it behaves exactly like myoglobin, but at four times the concentration. In reality however, the rate constants are far from constant, instead hemoglobin has the interesting property that it binds oxygen more readily the more oxygen it already has bound. This is known as *cooperative binding*, as the binding sites seem to be cooperating with each other.

The net results of cooperative binding is that hemoglobin will tend to be found at either extremes, either with no oxygen bound, or all four oxygen bound, i.e., an all-or-nothing binding. To model this binding let us make the simplifying assumption that hemoglobin binds all four oxygen molecules simultaneously. This assumption isn't strictly true, but it is a very useful approximation. 

<img src="fig/hemoglobin_reaction_simplified.png" width=200>

<center><b>Figure 3b:</b> Due to cooperative binding, hemoglobin tends to bind oxygen in an all-or-nothing fashion. We can therefore approximate the reaction as a single step.</center>

Under this approximation the reaction simplifies to a single reversible reaction, described by the following chemical equation

\begin{equation}
\mathrm{Hb} + 4\mathrm{O_2} \underset{k_-}{\overset{k_+}{\rightleftharpoons}} \mathrm{Hb(O_2)_4}.
\end{equation}

We can again apply the law of mass action to find the rate of this reaction. Because four oxygen molecules are now to bind simultaneously, the rate will be proportional to the concentration of all five reactants, meaning the reaction rate will be proportional to the oxygen concentration to the fourth power

\begin{equation}
\text{forward rate} \propto [\mathrm{Hb}] \times [\mathrm{O}_2] \times [\mathrm{O}_2] \times [\mathrm{O}_2] \times [\mathrm{O}_2] = [\mathrm{Hb}][\mathrm{O}_2]^4.
\end{equation}

Put simply, having the reaction depend on four oxygen molecules makes the rate dependence on the oxygen concentration a lot steeper. Writing out the full forward and reverse rates gives the change in oxygen-bound hemoglobin to be 

\begin{equation}
\frac{\mathrm{d[Hb(O}_2)_4]}{\mathrm{d} t} = k_+ \mathrm{[Hb]}\mathrm{[O_2]}^4 - k_- [\mathrm{Hb(O}_2)_4].
\end{equation}

Which in turn yields the same equilibrium equation as earlier, but with the oxygen concentration taken to the fourth power

\begin{equation}
\frac{\mathrm{[Hb]}\mathrm{[O_2]}^4}{\mathrm{[Hb(O_2)_4]}} = K_{\rm d}.
\end{equation}

The derivation is so far very similar to the one we made for myoglobin, and the only difference is in the power of four. However, it is no immediately apparent how this exponent affects the system in practical terms. Let us therefore derive the fractional degree of oxygenation of hemoglobin, so we can compare this to the one we found for myoglobin. Finding this expression is practically identical to how we did it earlier, and it can be a good exercise for the reader to carry this out. You should find the following function

\begin{equation}
Y = \frac{\mathrm{[O_2]}^4}{\mathrm{[O_2]}^4 + K_d}.
\end{equation}

### Comparing the fractional binding curves of myoglobin and hemoglobin

If you compare the expressions we found for the fractional degree of oxygenation for myoglobin and hemoglobin, the only difference is the exponent on the concentration of free oxygen. Both equations are on the form 

\begin{equation}
Y = \frac{\mathrm{[O_2]}^n}{\mathrm{[O_2]}^n + K_d},
\end{equation}

This equation is a specific example of the *Hill equation*, and the parameter $n$ is typically called the Hill coefficient. Thus, oxygen binding to myoglobin has a hill coefficient of $n=1$, while for hemoglobin it is $n=4$.

Note also that some care should be taken when discussing the dissociation constant $K_{\rm d}$. In our case, we introduced it as $k_{\rm -}/k_{\rm +}$, but for our expression of $Y$ to make sense, notice that $K_{\rm d}$ has to have units of mol$^{n}$. For myoglobin this is simply units of concentration, but for hemoglobin it is a lot less intuitive. Some therefore prefer to define a quantity $K_{\rm A}$, which obeys

\begin{equation}
(K_{\rm A})^n = K_{\rm d}.
\end{equation}

This quantity is usually referred to as the *microscopic* dissociation constant, while $K_{\rm d}$ can be referred to as the *apparent* dissociation constant. From the definition $K_{\rm A}$ will have units of concentration and it is when $[\mathrm{O}_2] = K_{\rm A}$ that the binding will be half-saturated.

Again, there are no strict conventions across the literature, and some refer to the microscopic dissociation constant as $K_{\rm d}$, which can definitely lead to confusion. Because of the varying conventions, we recommend you be extra mindful of units when modeling chemical reactions and ensure you are using the appropriate parameters.

To better understand the impact of the difference in the Hill coefficient between the two functions, let us plot the two curves in the same figure, for the same half-saturation point $K_{\rm A}$. This graph is shown below. We see that by going from a Hill coefficient of 1 to 4 makes the function a sigmoidal S-shape with a much steeper rise. The two curves intersect in the half-saturation point (because they have the same $K_{\rm A}$), but the curve for $n=4$ goes towards zero much faster below this point, and goes to one much faster above it.

<img src=fig/hill_equation.png width=400>

<center><b>Figure 4:</b> The Hill equation for $n=1$ and $n=4$ plotted in the same figure assuming the same microscopic dissocation constant $K_{\rm A}$.</center>

Figure 4 is good to show the effect of the Hill coefficient, given an equal half-saturation point. When comparing myoglobin and hemoglobin however, it would be misleading to use this figure, because the two have very different dissociation constants as well. The actual binding curves of myoglobin and hemoglobin are shown in figure 5, which shows quite a different picture. Here the oxygen concentration is shown in terms of *partial pressure* in units of mm Hg, which is a common clinical way of specifying oxygen levels in blood or tissue---the details are unimportant, you can simply think of this as a concentration of free oxygen.

<img src=fig/myoglobin_vs_hemoglobin.png width=400>

<center><b>Figure 4b:</b> The fractional degree of oxygenation for myoglobin and hemoglobin plotted against the partial pressure of oxygen, a common clinical way of measuring oxygen levels.</center>

In the lungs, the partial pressure of oxygen is on the order of 100 mm Hg, while in muscle tissue, it is about 20 mm Hg. If we compare the fractional binding curves of myoglobin and hemoglobin at these two levels, we see that the two reach about the same saturation in the lungs, but in the muscles, hemoglobin has a much lower fractional binding. Effectively, this means that hemoglobin is much better suited to act as an oxygen carrier, as it can pick up oxygen in the lungs, and deposit this in the muscles. Myoglobin on the other hand, does not act that well as an oxygen carrier, because it struggles to effectively part with its oxygen unless there is very little oxygen present.

The difference between myoglobin and hemoglobin make them suited for different tasks. Hemoglobin is found primarily in the blood, where it acts as an oxygen carrier, while myoglobin is found in tissue where it acts more like an oxygen reserve or storage. This functional difference between the two is a results of the cooperativity of the four hemoglobin subunits. This is a common theme in biological systems, where cooperativity between different units will make transitions steeper and functionally different. The Hill equation is therefore very general, and is useful for far more than modeling oxygen-binding.






### The general Hill equation

The Hill equation was derived in 1910 by Archibald Hill specifically to describe the binding curve of hemoglobin. The equation is however much more general than this specific example, and can be used to model many types of binding reactions. To emphasize this, let us write it out as a more general equation. In biochemistry, it is common to call the ion or molecule that is being bound the *ligand* and the macromolecule that binds it the *receptor*. A general ligand-receptor binding can then be described with the equation
\begin{equation}
\theta = \frac{[L]^n}{[L]^n + K_{\rm d}}.
\end{equation}
Here, $\theta$ is the fractional binding, $[L]$ the free ligand-concentration, $K_{\rm d}$ the dissociation constant of the binding reaction and $n$ the Hill coefficient.

### Interpreting the Hill coefficient

In our example, we showed that the binding curve of hemoglobin has a Hill coefficient of 4, because it binds four oxygen molecules in an all-or-nothing fashion. Recall however, that this was an approximation we made. If you were to measure the actual oxygen binding curve of hemoglobin you would get a result that fits very well to the Hill curve, but with a Hill coefficient of $n=3$ rather than 4. This is because hemoglobin can, and will, bind one, two or three oxygen molecules. Therefore, the Hill coefficient shouldn't be interpreted strictly as the number of units cooperating, but rather the *degree* of cooperativity. In fact, $n$ can even be fractional. Depending on the exact conditions under which you measure the oxygen binding, you'll find a best fit using a coefficient in the range 2.8-3.1. If we interpret the coefficient as a phenomenological parameter trying to catch the degree of cooperativity, using a fractional coefficient to get the best possible fit to measured data makes perfect sense.

In general, we can say that $n = 1$ means no cooperativity, i.e., independent binding sites. When we have 
cooperative binding, we have $n > 1$. Note also that one can have *negative cooperation*, meaning binding to one site inhibits binding to other sites. In such a case we would have a coefficient $n < 1$.

# Gibbs Free Energy

When analyzing reactions and processes occurring at the cellular level it can be very useful to consider the fundamental thermodynamic quantities of the system, namely energy and *entropy*. This can tell us when a given process will occur spontaneously and give us a better understanding of equilibrium dynamics. A very useful concept in such analysis is *Gibbs energy*, which combines energy and entropy into a single quantity.

The first law of thermodynamics is a statement about conservation of energy and the total energy of an isolated system will always be constant over time. A reacting system in the cell however, is not isolated and it can exchange energy with its environment through either heat exchange or mechanical work. A given process can therefore either absorb energy from its environment or deposit it.

The second law of thermodynamics states that the total entropy of an isolated system will never decrease, it can only increase over time. Put simply, entropy is a measure of a systems disorder and is a physical property much like energy. Like energy, the entropy of the reactants in a process does not necessarily have to increase, but the net entropy of the reactants and their environment combined can only increase in a spontaneous reaction. An isolated system will spontaneously evolve towards its maximum possible entropy, which will correspond the the thermodynamic equilibrium state of the system. Put simply, it is the increase in entropy over time that is the driving force of chemical reactions and cellular processes.

To properly understand the thermodynamics of a process, we need to analyze how both energy and entropy changes in the reactants and their environment. We can combine these aspects into a single quantity, called the *Gibbs free energy* or simply *Gibbs energy* of the system. It can be expressed as 
\begin{equation}
G = U + pV - TS.
\end{equation}
Here, $G$ is the Gibbs energy of the system, $U$, $V$ and $S$ are the systems internal energy, volume and entropy respectively. Lastly we have the pressure, $p$, and temperature, $T$, of the surrounding environment.

From this definition, $G$ will have units of energy, which is why we call it Gibbs *energy*, but note that entropy is also baked into this definition. With the introduction of Gibbs energy we can restate the second law of thermodynamics not in terms of entropy, but in terms of Gibbs energy. The net Gibbs energy of any isolated system cannot increase, but only decrease. An isolated system will spontaneously tend towards its minimal Gibbs energy, which will correspond to its thermodynamic equilibrium.


### Why is it called Gibbs *free* energy?

Gibbs energy is known by many names. Some, like us, call it just Gibbs energy, while it is also commonly called Gibbs \emph{free} energy. Other names are occasionally used, such as Gibbs function or free enthalpy. When J. W. Gibbs introduced the concept in the 1870s he dubbed it simply *available energy*. This is because the Gibbs energy can be thought of as the maximum amount of work that can be extracted as useful work from a given system. 

A good metaphor for Gibbs energy is found in Schroeder's *An Introduction to Thermal Physics*. Schroeder asks you to imagine a wizard conjuring up a rabbit from nothing. Obviously, the wizard has to respect the laws of thermodynamics, which state that energy cannot be created, and so he will have to supply the needed energy in the process. But how much energy would the wizard need to supply to create the rabbit? First he has to produce the internal energy of the rabbit itself, $U$, which encompasses all the mass energy and all internal chemical energy and so on. If the rabbit were to be created in a complete vacuum, this would be the end of it. However, to get the rabbit to pop into existence in front of you, the wizard actually has to make space for the rabbit by pushing some air aside, which costs mechanical energy. This cost would be $pV$, the higher the air pressure or the bigger the rabbit, the more work has to be performed. The energy bill is now $U + pV$. However, the wizard is quite clever, and so he can save some energy as well. If he conjures up a cold rabbit, he knows the laws of thermodynamics says heat will spontaneously flow from the high temperature surroundings into the low temperature rabbit, so the wizard can subtract this energy from his energy bill. This contribution is $TS$, where $T$ is the temperature of the surroundings and $S$ is the entropy of the rabbit. Why does the entropy enter into it? Because temperature is defined as derivative $T = \delta U/\delta S$. So the more entropy the rabbit has, the more thermal energy it will absorb from its environment. The total energy bill the wizard needs to pay is therefore the Gibbs energy $U + pV - TS$.

Schroeder's analogy nicely illustrates how Gibbs energy combines different thermodynamics considerations into a single quantity. But if the Gibbs energy represents a *cost*, why is it called the *free* energy? The name describes the reverse process. If an evil wizard were to annihilate a rabit and extract as much energy as possible in the process, we would first get the internal energy $U$ and some extra mechanical work from the surrounding air collapsing into the resulting vacuum. However, they would have to dump som of that energy as heat into the environment to satisfy the second law. The Gibbs free, or available, energy is thus simply the amount of energy a system could release if it were to cease existing.

Of course, in the real world a system doesn't simply cease existing. However, in any chemical reaction or other kind of process we can think of it as on system being annihilated and a new one being created. It is the *net change* in the Gibbs energy that is the important quantity.

### Change in Gibbs energy

In a given reaction, the energy, volume or entropy of the system can change. Any change in the pressure or temperature however, can be neglected, because we assume the surrounding environment to be much larger than the reacting system. Thus, the total change in Gibbs energy can generelly be written out as 

\begin{equation}
\Delta G = \Delta U + p\Delta V - T\Delta S.
\end{equation}

Thus the net change in Gibbs energy will be a weighted sum of its change in internal energy, $\Delta U$, its change in volume, $\Delta V$, or a change in net entropy $\Delta S$. 

Because the net Gibbs energy of any system will spontaneously decrease over time, any process or reaction where $\Delta G < 0$ will occur spontaneously. Any reaction where $\Delta G > 0$ cannot happen (it would instead occur in the reverse direction). If $\Delta G = 0$ we are at a thermodynamic equilibrium.

For any molecular species, we can express its Gibbs free energy as

\begin{equation}
G = G^0 + RT \ln c.
\end{equation}

Where the first term, $G^0$, is called the standard free energy, which is dependent on what chemical species is being discussed, but not the concentrations involved. The second term is dependent on the concentration of the species, but is otherwise inconsiderate of what the species is. Here, $R$ is the gas constant, $T$ the temperature and $c$ the concentration of the species.

If a reaction occurs where A $\to$ B, the net change in Gibbs energy can be expressed as

\begin{equation}
\Delta G = \Delta G^0 + RT \ln \frac{c_{\rm B}}{c_{\rm A}}.
\end{equation}

Note that the change in Gibbs energy can be written as two terms. The first is the change in standard free energy, which depends on what molecular species are interacting. This value is tabulated for most molecular species under standard conditions. The second term only cares about the ratio of the concentrations of the species, and so is not specific to what reaction is occurring.

#### Chemical Equilibrium from Gibbs energy

As stated, we know that a system in equilibrium with regard to a reaction should have $\Delta G = 0$ for that reaction. So if we have system that can inhibit two given states, A and B, for example myoglobin and oxygen binding and releasing, then we have the equilibrium condition

\begin{equation}
\frac{c_{\rm B}}{c_{\rm A}} = e^{-\Delta G^0/RT}.
\end{equation}

Here the left hand side of the equation is the ratio of the two concentrations, while the right hand side of the equation are constants. While this equilibrium condition looks very different from the one we found when discussion the law of mass action, it effectively states the same thing. This means that if we know the change in standard free energy for some reaction or process, we can find the concentration ratio at equilibrium.

### The Cell Membrane

The structure of the cell membrane itself is fairly complex, but its fundamental building blocks are molecules known as phospholipids, which are quite simple. A sketch of such a molecule is shown in the figure below, and the important takeaway is that one side of the molecule is hydrophobic and the other is hydrophilic. 

<img src=fig/phospholipid.png width=300>

<center><b>Figure 6:</b> The basic building block of the cell membrane is the phospholipid, a simple molecule consists of a hydrophilic phosphate head, and a hydrophobic tail built up of two fatty acid chains. </center>

Hydrophobic, or non-polar, molecules do not mix well with water and tend instead to separate, like when mixing oil and water. This happens because non-polar molecules disrupt hydrogen bonds between water molecules, which incurs an increase in the Gibbs energy of the system. By clumping together this effect is minimized and therefore the mixture will spontaneously separate to minimize its Gibbs energy. This is known as the *hydrophobic interaction*. The situation is reversed for hydrophilic, or polar, molecules. Distributing these evenly in a solution increases the entropy of the system and therefore reduces the net Gibbs energy of the system. Polar molecules will therefore spontaneously dissolve into a solution.

Phospholipids are *both* hydrophilic and hydrophobic at the same time, so what happens when we try to dissolve them into water? Ideally the heads would dissolve while the tails clump together. Because the two are stuck together they must compromise, and they therefore self-assemble into a structure where the hydrophobic tails are shielded from the water while the hydrophilic heads face outward and towards the surrouding water molecules. One such structure is the *lipid bilayer*.

In the bilayer, the phospholipids create two planar sheets that lie in parallel, with the tails of each pointing towards the center. The layers do not allow water to pass in, and so the structure very effectively hides the hydrophobic tails away from the water. The bilayer will have a fixed thickness, but there is no real limit to how large the sheet itself can be, given enough phospholipid molecules. The sheet itself is also quite flexible, and so the whole sheet can loop back around and meet itself, giving a balloon structure. The membrane of a cell is precisely such a structure.

<img src="fig/lipid_bilayer.png" width=400>

<center><b>Figure 6b:</b> A lipid bilayer is formed by two parallel planar sheets of phospholipids such that the heads are on the outside and the tails are inside the structure and therefore shielded from water molecules.</center>

The bilayer has a fixed thickness and it is very thin, about 5 nm across. However, the sheet itself can grow if more and more phospholipids are added. The sheet itself is also quite flexible, and as it becomes large enough it can be curved around and meet itself, forming a balloon-like structure. A typical cardiomyocyte is cylindrical and around 100 µm long and and about 20 µm in diameter and has a lipid bilayer as its cell membrane separating the *intracellular* and *extracellular* domains, i.e., the inside and outside of the cell. Another name for the inside of the cell is the *cytosol* and so the term *cytosolic* refers to seomthing inside the cell.

### Permeability of the membrane

The bilayer of the membrane consists of phospholipids that self-assemble to avoid hydrophobic interactions. Note however that there are no explicit forces holding the lipid molecules together and they are therefore more or less free to float and drift in relation to each other as long as they remain within the bilayer plane. The bilayer is therefore a dynamic and fluid substance and it is possible for other molecules to slip in between the lipids and move into or through the lipid bilayer. Smaller molecules can therefore diffuse across the cell membrane. However, the membrane does not treat all molecules equally.

Due to the nature of the lipid bilayer, the interior of the membrane is a non-polar environment, and polar molecules will therefore prefer to stay in the polar environment outside the membrane, rather than move through the non-polar membrane. Non-polar molecules however, such as oxygen or carbon dioxide, have little issue diffusing through the membrane.

Note that all charged particles are polar due to their electric charge and the membrane is very effective at stopping such charged particles from passing through. The membrane is thus impermeable to charged ions, and because we cannot have any electric current without a charge carrier, the membrane itself acts like an electrical insulator.

### The Cell Membrane as a Capacitor

The lipid bilayer itself is impermeable to charged ions and acts like an electrical insulator. In comparison the inside and outside of the cell have many dissolved charged ions that can carry a current and are therefore good electric conductors. When two electric conductors are separated by an insulator, we get what we call an electric *capacitor*.

If an electric potential is applied across a capacitor, an electric current *wants* to flow through, but the electric insulator prevents this, so instead we get a pileup of electric charge on both sides of the insulator. This build-up continues until these charges set up an electric field that is of equal strength to the applied electric potential. Effectively the applied potential charges the capacitor like a small battery and once charged it perfectly balances out the imposed electric potential.

The amount of charge that builds up on either side of the insulator in a capacitor at equilibrium is proportional to the potential across it
\begin{equation}
V = \frac{Q}{C_{\rm m}}.
\end{equation}
Here $V$ is the potential, or *voltage*, across the capacitor and $Q$ is the total charge separated by it. The proportionality factor is called the capacitance, which is a physical property of the capacitor in question. 

As we are discussing the cell membrane as a capacitor, the capacitance in question is the *membrane capacitance* and we therefore denote it $C_{\rm m}$. The capacitance of the lipid bilayer is about 1 µF/cm$^2$, where F is the unit of capacitance, Farad. If we multiply this quantity with the full surface of the cell membrane, the whole cell has a capacitance of about 0.05 nF. This might seem like a very small quantity because it is given in terms of nanofarads, however, note that Farad is an absolutely huge unit. Most capacitances seen in electrical components are in the range of pF to µF. At the same time, the fairly tiny size of the cell means that membrane capacitance is actually fairly high.

The variable $V$ represents the electric potential difference across the membrane, which we often just call the *membrane potential*. It is this variable we are most interested in modeling when it comes to cell electrophysiology.

<center><b>Figure 7:</b> The intracellular and extracellular spaces are separated by the cell membrane, which is impermeable to charged particles and acts like an insulator. If we apply an electric field across the membrane, it will act as an electrical capacitor and there will be a build-up of charge on both either side of the membrane.</center>

We are interested in modeling how the membrane potential $V$ changes over time. Any time there is a change in the charge separation across the membrane, the electric field across it will change and the membrane potential will shift so that it again cancels out the field. Because electric phenomenon are generally very fast, we can assume the membrane as a capacitor to always be at its equilibrium potential, i.e., we assume quasi-static conditions.

If we assume the proportionality equation to hold at all times, we can take the derivative to find the differential equation
$$\frac{{\rm d} V}{{\rm d}t} = \frac{1}{C_{\rm m}} \frac{{\rm d} Q}{{\rm d} t}.$$
The time derivative of a charge can be interpreted as a current, and we call this the *capacitive current*, or $I_{\rm cap}$. The capacitive current is called a *virtual* current, as it isn't an actual carrier mediated current carried across the insulator (which cannot happen), but rather a current that comes from the charge build up or release of built up charge. The equation can therefore be written 
$$\frac{{\rm d} V}{{\rm d}t} = \frac{1}{C_{\rm m}} I_{\rm cap}.$$
This is the differential equation we will use to model the change in membrane potential over time. But to solve it we first need to find an expression for the capacitive current $I_{\rm cap}$.

To summarize, the cell membrane itself can be thought of as an electrical capacitor separating the inside and outside of the cell and we can simplify this situation with the following electric circuit diagram. We denote the intra- and extracellular potentials as $V_{\rm i}$ and $V_{\rm e}$, the membrane potential is the difference between these, but can be defined in either direction. In practice it is most common to define it as 

$$\Delta V = V_{\rm i} - V_{\rm e}.$$

Even though the membrane potential is the potential difference, we denote it as simply as $V$, forgoing the $\Delta
$-symbol for neater notation.

### Ion Channels
The cell membrane does not only consist of the lipid bilayer, but also a large number of other molecules and proteins embedded in the bilayer. ### Passive and active ion transport

Ion channels create a pore in the membrane that allows charged ion to pass through, but the movement of ions through this channel has to occur spontaneously due to either diffusion or an electric force pushing them through. This is known as *passive* ion transport. The membrane also contains molecular machinery known as *ion pumps* and *exchangers* which actively move ions across the membrane. 

The important difference between passive and active ion transport is that passive transport has to occur spontaneously and so must satisfy $\Delta G < 0$. While active transport is coupled to some external energy source and so can give transport against this gradient. A useful analogy for passive ion transport would be a staircase, which allows people to move between floors but they still have to walk between the floors. Here active transport would be like an elevator.

For the rest of this lecture and in the coming days we will mostly be concerned with modeling passive ion transport, as this type is much more important for modeling the action potential. We will return to the exchangers and pumps later.

# Electrodiffusion with the Nernst-Planck equation

When an ion channel is open, ions are free to move either into or out of the cell. Motion of individual ions at the microscopic scale is random and chaotic due to thermal motion, but we can still model the *net* motion of ions, which is the macroscopic quantity which will have an impact on the membrane potential.

As mentioned, transport through channels is *passive*, meaning the ions move on their own accord. There are two different driving forces that affects the net motion of ions, and these are *diffusion* and *electric drift*. The combination of these two driving forces is usually referred to as *electrodiffusion*.

#### Diffusion and Fick's law

Diffusion describes the process were concentration differences tend to smooth out over time. If a system has a non-uniform concentration of a given ion or molecule, then random thermal motion will tend to erase these differences over time. At the macroscopic level this will look like there is a net *diffusive flux* of mass going against the concentration gradient. This can be described by Fick's law of diffusion (technically Fick's *first* law of diffusion) which states that the diffusive flux is proportional to the concentration gradient. In 1D it can be expressed as 

\begin{equation}
J(x) = -D \frac{\partial c(x)}{\partial x},
\end{equation}

while in higher dimensions we would state it on vector form as

\begin{equation}
\vec{J}(\vec{r}) = -D \nabla c(\vec{r}).
\end{equation}
Here, $J$ is the net diffusive flux, $c$ is the concentration and $D$ the proportionality constant, called the *diffusivity* or *diffusion coefficient*. Note that while both $J$ and $c$ are field variables, meaning they have different values in different locations $\vec{r}$, we will from now on forgo writing their spatial arguments to get a neater notation.

Note that Fick's law holds true for each ionic or molecular species independently, so it is often written with a subindex denoting specific species, or generally as 
\begin{equation}
\vec{J}_k = -D_k \nabla c_k.
\end{equation}
where $k$ denotes the species. Note that the diffusivity is also species dependent.

What Fick's law tells us is that concentrations will spontaneously flow to level out concentration differences. If the concentration of a given ionic species is higher outside the cell than inside it, then there will be a diffusive flux pointing into the cell, and vice versa. If the concentration is the same on either side, there will still bed random thermal motion of ions into and out of the cell, but there will be no effective *net* motion due to diffusion.

#### Electric Drift of Ions

When subject to an electric field charged particles will begin to move. This will lead to a flux of matter. The flux will be proportional to the strength of the electric field and to the charge of the particle. We have the flux

\begin{equation}
J_{\rm drift} = \frac{DzF}{RT}c\vec{E}.
\end{equation}
Here $D$ is again the diffusivity, $z$ is the *valency* of the species in question, i.e., their electric charge, while $F$, $R$ and $T$ are the Faraday constant, the gas constant and the temperature. Finally we have the electric field $\vec{E}$.

Any electric field can be described as the gradient of an electric potential, and this is useful because it is the membrane *potential* which we are most interested in modeling, we therefore use the expression $\vec{E} = -\nabla V$ to rewrite this equation as

\begin{equation}
J_{\rm drift} = -\frac{DzF}{RT}c\nabla V.
\end{equation}

This equation tells us that if there is a potential $V$ across the membrane, then there will be a flux of charged ions through ion channels in the membrane due to electric forces driving ions through.

## The Nernst-Plank Equation

The *net* flux of ions through an ion channel will be given by the combination of diffusion and electrical drift, so the total flux can be written out as 

\begin{equation}
J = J_{\rm diff} + J_{\rm drift} = -D\left(\nabla c + \frac{zF}{RT}c\nabla V\right).
\end{equation}

This equation is known as the Nernst-Planck equation, and it describes the net flux of ions through an ion channels due to electrodiffusion. The two terms correspond to diffusion driven by concentration difference across the membrane, and drift driven by the potential difference across the membrane. This equation again holds for each species independently, and so we should technically write it out with the subindex $k$.

The Nernst-Planck equation can now be used to find the flux of electrically charged ions through the membrane, which in turn allows us to model how the membrane potential changes over time. To do that there is one last derivation we need to carry out, namely finding out the equilibrium potential of the membrane.

# The Nernst Potential as an equilibrium

We have now derived the Nernst-Planck equation which states that passive transport of ions through the membrane is driven by the sum of diffusion and electric forces. Note that these two driving forces are in general independent of each other and can point in the same, or in opposite directions. As charged ions move through the membrane they will carry an electrical current which will shift the membrane potential. We will now see how this leads to a situation where the system tends towards an equilibrium state were the diffusive and electric terms in the Nernst-Planck equation cancel each other out.

We now create a little thought experiment. Assume that we have a mixture of positively charged K$^+$-ions and negatively charged $Cl^-$-ions. There are equal amounts, so the mixture is electrically neutral. Now we place a high concentration of the mixture inside the cell, and a low concentration outside the cell. Both sides of the membrane are now electrically neutral, as a result, there will be no membrane potential ($V=0$). However, because the concentration is higher inside than outside there will be a concentration gradient across the membrane.

Assume now that there are potassium specific ion channels in the membrane that now open up, meaning K$^+$-ions are free to move into or out of the cell. Just as the channels open, there is a large concentration gradient out of the cell but no electric potential. This means that the Nernst-Planck equation reduces to Fick's law and potassium ions will start to diffuse out of the cell through the channels. When potassium ions leave the cell they leave behind a surplus of negatively charged Cl$^-$-ions that have no channels they can diffuse through. This effectively means that as K$^+$-ions leave they cell the cell becomes electrically charged.

As a result of the electrical imbalance, an electric field will start to form across the membrane which starts to inhibit K$^+$-ions from leaving the cell. Eventually the membrane potential becomes strong enough to cancel out the diffusive flux entirely and the system has reached a new equilibrium. There will still be a concentration gradient across the membrane as well as a membrane potential, but these two perfectly cancel each other out. The situation is shown in the figure below.

* **Center:** Potassium channels in the membrane open, potassium starts to diffuse out of the cell as a result of the concentration gradient.

### Deriving the Nernst Potential

At what potential does the net transport of ions through the channels stop? We can use the Nernst-Planck equation to find this equilibrium value. Because we are finding an equilibrium, there must be zero net flux, so we get
$$J = -D\bigg(\nabla c + \frac{zF}{RT}c\nabla V\bigg) = 0.$$
We first note that we can discard the diffusion coefficient entirely, apparently, the rate at which diffusion happens won't affect the conditions for *if* diffusion happens. This leaves
$$\nabla c = \frac{zF}{RT}c\nabla V.$$
To make sense of this equation, let us make a rough sketch of the membrane and the concentration and electric gradients. Assuming the outside and inside are well mixed, there are only gradients across the membrane itself.

<img src="fig/nernst.png" width=400>

The gradients are pointed straight through the membrane, so we can turn the equation one-dimensional
$$\frac{{\rm d} c}{{\rm d} x} = \frac{zF}{RT}c\frac{{\rm d} V}{{\rm d} x}.$$
To find an expression for the potential $V$, we integrate this equation across the membrane
$$\int_0^L \frac{1}{c} \frac{{\rm d} c}{{\rm d} x} \ {\rm d} x = \frac{zF}{RT} \int_0^L  \frac{{\rm d} V}{{\rm d} x} \ {\rm d} x.$$
With a substitution of variables
$$\int_0^L \frac{1}{c} \ {\rm\ d} c = \frac{zF}{RT} \int_0^L  {\rm d} V.$$
Which gives
$$\ln c_{\rm e} - \ln c_{\rm i} = \frac{zF}{RT}(V_{\rm e} - V_{\rm i}).$$
And so the equilibrium condition can be summarized as
$$V_{\rm eq} = \frac{RT}{zF} \ln \frac{c_{\rm e}}{c_{\rm i}}.$$
At this specific membrane potential (for a given concentration difference), the electric and diffusion forces on the ions perfectly balance, giving no net transport of ions across the membrane.

This equilibrium potential is referred to as the *Nernst potential*. It is specific to a given ion species, as it depends on the ionic concentrations $c_i$ and $c_e$, and the valency $z$. It is therefore often denoted as $E_{\rm X}$ for ionic species X. So for example the Nernst potential for sodium, potassium and calcium would be: $E_{\rm Na}$, $E_{\rm K}$ and $E_{\rm Ca}$ respectively.

### Finding the Nernst Potential via Gibbs Energy

An even simpler way of finding an expression for the Nernst potential is to use the fact that there is no change in Gibb's energy for a process at equilibrium ($\Delta G = 0$). Thus, the process of moving an ion across the membrane should not incur any change in the Gibb's energy, given that $V = E_{\rm X}$.

The Gibb's energy for a particle inside the cell is
$$G_{\rm i} = G^0 + RT\ln [{\rm X}]_{\rm i} + zFV_{\rm i},$$
while on the outside it is
$$G_{\rm e} = G^0 + RT\ln [{\rm X}]_{\rm e} + zFV_{\rm e}.$$
When we combine the two, the standard free energy cancel out, giving the net change
$$\Delta G = G_{\rm i} - G_{\rm e} = RT \ln \frac{[{\rm X}]_{\rm i}}{[{\rm X}]_{\rm e}} + zFV.$$
And so, because $\Delta G = 0$ we get
$$zFV =  RT \ln \frac{[{\rm X}]_{\rm i}}{[{\rm e}]_{\rm i}},$$
or
$$V = \frac{RT}{zF}\ln \frac{[{\rm X}]_{\rm e}}{[{\rm X}]_{\rm i}} = E_{\rm X},$$
which is the Nernst potential.

While the two derivations of the Nernst potential might look very different, they both rely on the same underlying assumptions. And so it might be a good idea to become comfortable with both derivations.

## The Nernst Potential as a driving force

If the cell membrane is permeable to an ion type, and the membrane potential is different to the Nernst potential of that ion type, there will be a net transport of ions through the membrane. 

Let's take Na$^+$ as an example, if the membrane has open sodium channels, ions of this type can move through the membrane. If $V=E_{\rm Na}$, they have no desire to do so, but if $V \neq E_{\rm Na}$, there will be a net transport of sodium that tries to shift the membrane potential closer to the Nernst potential. This means that if $V < E_{\rm Na}$, sodium will enter into the cell (influx), while if $V > E_{\rm Na}$, sodium will tend to leave the cell, to shift the membrane potential downward (efflux). For this reason, the Nernst potential is also often called the *reversal potential*, because this is the potential at which a given flux of ions reverses its direction.

The bigger the difference in the actual membrane potential, and the equilibrium Nernst potential, the stronger the current will be. We refer to the difference $(V - E_{\rm Na})$ as the *driving force*. To find the actual current, we assume it to be Ohmic, meaning the magnitude of the current is directly proportional to the driving force
$$I_{\rm Na} = g_{\rm Na} (V - E_{\rm Na}).$$
The proportionality factor is the *conductance*, $g_{\rm Na}$, which is also species specific. The more open sodium channels there are in the membrane, the more ions can pass through per time, and the stronger the conductance and the stronger the current for a given driving force. The conductance is also the inverse of electrical resistance $g = 1/R$, so this is just a formulation of Ohm's law ($I = V/R$), but with the driving force being replaced by $V - E_{\rm Na}$, as we are now looking at electrodiffusion, and not just electric forces.

If $I_{\rm tot}$ was different from 0 there would be a net charge leaving the cell, which would not be sustainable over a longer time perspective, as charge separation requires large amounts of energy.

This means we can express the change of the membrane potential over time as
$$C_{\rm m}\frac{{\rm d}V}{{\rm d}t} = I_{\rm cap} = -I_{\rm Na}.$$
Which, when inserting for our Ohmic current gives
$$C_{\rm m}\frac{{\rm d}V}{{\rm d}t} = - g_{\rm Na} (V-E_{\rm Na}).$$

If we want to include currents of all three ionic species, then the total current is given as
$$I_{\rm tot} = I_{\rm cap} + I_{\rm ion} = I_{\rm cap} + I_{\rm Na} + I_{\rm K} + I_{\rm Ca} = 0.$$
Which gives the ODE
$$C_{\rm m}\frac{{\rm d}V}{{\rm d}t} = -I_{\rm ion} = -\big[g_{\rm Na} (V-E_{\rm Na}) + g_{\rm K} (V-E_{\rm K}) + g_{\rm Ca} (V-E_{\rm Ca})\big].$$

This is the ODE we solve to model the membrane potential over time.
